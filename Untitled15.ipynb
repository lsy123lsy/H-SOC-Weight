```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as la
# 
a = 2
# 
a1 = np.array([1/2, 1/2, 0]) * a
a2 = np.array([0, 1/2, 1/2]) * a
a3 = np.array([1/2, 0, 1/2]) * a
neighbors = [a1, a2, a3]

# eV
V_ss = -0.6
V_sp = -0.4
V_pp_sigma = 0.7  # 
V_pp_pi = 0.6  # 

epsilon_s = 0.8  # seV
epsilon_p = 1.0  # peV

Ess = V_ss    # Ess = Vss_sigma
Esp = V_sp/np.sqrt(2)    # Esp = lVsp_sigma,lx45l1/sqrt(2)
Exx = Eyy = Ezz = V_pp_sigma/2 + V_pp_pi/2    # Exx = l**2(Vpp_sigma) + (1-l**2)Vpp_pi45l1/sqrt(2)
Exy = Exz = 0    # Exy = lmVpp_sigma - lmVpp_pi, xpi/2yzpi/4


# k
points = np.array([
    [3*np.pi/(4*a), 3*np.pi/(4*a), 3*np.pi/(2*a)],  # K
    [0, 0, 0],  # G
    [np.pi/a, np.pi/a, np.pi/a],  # L
    [np.pi/a, np.pi/(2*a), 3*np.pi/(2*a)],  # W
    [np.pi/a, 0, np.pi/a]  # X
])

# 
def compute_lengths(points):
    lengths = [0]
    step = []
    for i in range(1, len(points)):
        length = np.linalg.norm(points[i] - points[i-1])
        lengths.append(lengths[-1] + length)
        step.append(length)
    lengths = np.round(lengths, 3)
    step = np.round(step, 3)
    return lengths, step

# k
def k_path(lengths, step, nk=100):
    G = np.array([0, 0, 0])
    X = np.array([np.pi/a, 0, np.pi/a])
    W = np.array([np.pi/a, np.pi/(2*a), 3*np.pi/(2*a)])
    L = np.array([np.pi/a, np.pi/a, np.pi/a])
    K = np.array([3*np.pi/(4*a), 3*np.pi/(4*a), 3*np.pi/(2*a)])
    path = np.vstack([
        np.linspace(K, G, int(nk*step[0]/lengths[-1])),
        np.linspace(G, L, int(nk*step[1]/lengths[-1])),
        np.linspace(L, W, int(nk*step[2]/lengths[-1])),
        np.linspace(W, X, int(nk*step[3]/lengths[-1]))
    ])
    return path

def H_k(k):
    kx, ky, kz = k
    
    H = np.zeros((8, 8), dtype=complex)
    
    # 
    H[0, 0] = epsilon_s + 2*Ess*(np.cos(kx) + np.cos(ky) + np.cos(kz))
    H[1, 1] = epsilon_p + 2*Exx*np.cos(kx)
    H[2, 2] = epsilon_p + 2*Exx*np.cos(ky)
    H[3, 3] = epsilon_p + 2*Exx*np.cos(kz)
    H[4, 4] = epsilon_s + 2*Ess*(np.cos(kx) + np.cos(ky) + np.cos(kz))
    H[5, 5] = epsilon_p + 2*Exx*np.cos(kx)
    H[6, 6] = epsilon_p + 2*Exx*np.cos(ky)
    H[7, 7] = epsilon_p + 2*Exx*np.cos(kz)

    H[0, 1] = 2j*Esp*np.sin(kx)
    H[0, 2] = 2j*Esp*np.sin(ky)
    H[0, 3] = 2j*Esp*np.sin(kz)
    H[1, 0] = np.conj(H[0, 1])
    H[2, 0] = np.conj(H[0, 2])
    H[3, 0] = np.conj(H[0, 3])
    H[4, 5] = 2j*Esp*np.sin(kx)
    H[4, 6] = 2j*Esp*np.sin(ky)
    H[4, 7] = 2j*Esp*np.sin(kz)
    H[5, 4] = np.conj(H[4, 5])
    H[6, 4] = np.conj(H[4, 6])
    H[7, 4] = np.conj(H[4, 7])

    # 
    # s 
    Lx_s = np.array([[0]])
    Ly_s = np.array([[0]])
    Lz_s = np.array([[0]])

    # p 
    Lx_p = np.array([[0, 0, 0],
                     [0, 0, -1j],
                     [0, 1j, 0]])

    Ly_p = np.array([[0, 0, 1j],
                 [0, 0, 0],
                 [-1j, 0, 0]])

    Lz_p = np.array([[0, -1j, 0],
                     [1j, 0, 0],
                     [0, 0, 0]])

    # 
    Sx = np.array([[0, 0.5],
                   [0.5, 0]])

    Sy = np.array([[0, -0.5j],
                   [0.5j, 0]])

    Sz = np.array([[0.5, 0],
                   [0, -0.5]])

    #  ()
    xi = 1.0

    # 
    LS_x_s = np.kron(Lx_s, Sx)
    LS_y_s = np.kron(Ly_s, Sy)
    LS_z_s = np.kron(Lz_s, Sz)

    LS_x_p = np.kron(Lx_p, Sx)
    LS_y_p = np.kron(Ly_p, Sy)
    LS_z_p = np.kron(Lz_p, Sz)

    #  s  p 
    LS_x = la.block_diag(LS_x_s, LS_x_p)
    LS_y = la.block_diag(LS_y_s, LS_y_p)
    LS_z = la.block_diag(LS_z_s, LS_z_p)

    H_SO = xi * (LS_x + LS_y + LS_z)

    H += H_SO
   
    return H

# 
def calculate_bands_and_weights(path):
    bands = []
    weights = []
    for k in path:
        H = H_k(k)
        eigenvalues, eigenvectors = np.linalg.eigh(H)
        bands.append(eigenvalues)
        weights.append(np.abs(eigenvectors[7, :])**2)  # 4
    return np.array(bands), np.array(weights)

# 
def plot_bands_and_weights(lengths, bands, weights, path_length):
    plt.figure(figsize=(10, 8))
    for i in range(bands.shape[1]):
        plt.plot(bands[:, i], label=f'Band {i+1}')
        
    # 
    orbitals = ['s-u', 's-d', 'px-u', 'px-d', 'py-u', 'py-u', 'pz-u','pz-d']
    for i in range(weights.shape[1]):
        plt.scatter(range(len(weights[:, i])), bands[:, i], s=weights[:, i]*100, alpha=0.5, label=f'{orbitals[i]} orbital')
    
    plt.xticks([path_length*lengths[i] // lengths[-1] for i in range(len(lengths))], labels=['K', '', 'L', 'W', 'X'])
    plt.xlabel('PATH', fontsize=25)
    plt.ylabel('Energy (eV)', fontsize=25)
    plt.title('Band Str of FCC-SOC with pz-d-Weights', fontsize=25)
    plt.grid(True)
    plt.xlim(0, 97)
    plt.tick_params(labelsize=30)
    #plt.legend()
    plt.show()

# 
lengths, step = compute_lengths(points)
path = k_path(lengths, step)
bands, weights = calculate_bands_and_weights(path)
plot_bands_and_weights(lengths, bands, weights, len(path))

```
